# First Import Python Libraries needed

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import sklearn.linear_model as linear_model
import warnings
```

# Syrian analysis

```python
# first load the excel sheet for Syrian
syria = pd.read_excel("tabela_finale_Syria.xlsx")

#Choose Row and columns to keep to get rid of the empyty ones
syria = syria.loc[slice(0, 13), slice('territory', "Stool % 2013 WHO*")].fillna(0)

#Multiply the percentage by 100 to get it out of 100 not 1
syria.govr = syria.govr*100
syria['non-govnmt'] *= 100
syria
```

```python

# Choose Independent variables
x_syria = [
           'distance from capital in miles',
#            'govr',
           'non-govnmt'#,
#            'Hosted IDPs per governorate (June 2014, source: OCHA, UN)'
          ]

x_syria_comp = [
     'distance from capital in miles',
           'govr',
           'non-govnmt',
           'Hosted IDPs per governorate (June 2014, source: OCHA, UN)'
]

#Choose dependent variable
y_syria = 'confirmed cases (GPEI) + compatible cases (WHO) + compatible cases (ACU)'
```

# Linear Regression for Syria 

```python
# assign the independent variables to variable X with adding the constant for intercept computation
X = sm.add_constant(syria[x_syria])

#assign the dependent variables to variable y
y = syria[y_syria]#.values

#Peform the Linear regression using statsmodels library (named here as sm)
model = sm.OLS(y, X)
result = model.fit()

with warnings.catch_warnings():
    warnings.filterwarnings("ignore")
    print(result.summary())
```

## Using GLM with Poisson function

```python
# assign the independent variables to variable X with adding the constant for intercept computation
X = sm.add_constant(syria[x_syria_comp])

#assign the dependent variables to variable y
y = syria[y_syria]#.values

#Peform the Linear regression using statsmodels library (named here as sm)
model = sm.GLM(y, X, family=sm.families.Poisson())
result = model.fit()
print(result.summary())
```
